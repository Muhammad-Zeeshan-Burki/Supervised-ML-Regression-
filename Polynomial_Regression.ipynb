```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
dataset =pd.read_csv("synthetic-Modified.csv")
```

```python
dataset.head(3)
```

```python
dataset.corr()
```

```python
plt.scatter(dataset["Data"],dataset["Result"])
plt.xlabel("Data")
plt.ylabel("Result")
plt.show()
```

```python
x = dataset[["Data"]]
y =dataset[["Result"]]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
Pf =PolynomialFeatures(degree=2)
```

```python
Pf.fit(x)
x= Pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test ,y_train,y_test =r=train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr =LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
prd =lr.predict(x)
```

```python
plt.scatter(dataset["Data"],dataset["Result"])
plt.plot(dataset["Data"],prd,c="red")

plt.xlabel("Data")
plt.ylabel("Result")
plt.legend(["org","Pridict Line"])

plt.show()
```
